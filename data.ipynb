```python
import requests
api = "http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp"
param = {'stnId' : '109'}
result = requests.get(api, params=param)
html = ""
if result.status_code == 200:
    html = result.text
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(html,"html.parser")
print(soup)
```

```python
all_datas = soup.find_all("data")
print(all_datas)
```

```python
time = soup.find_all("tmef")
tem = soup.find_all("tmn")

for t, em in zip(time, tem):
    print(t.get_text(), em.get_text())
```
